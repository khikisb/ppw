# Tugas 3 implementasi Latent Directlet allocation (LDA)

Lakukan implementasi kan lda pada data abstrack:

dengan k=3  alhpa=0.1 beta=0.2

library di scikit-learn

doc_topic_prior float, default=None

Prior of document topic distribution theta. If the value is None, defaults to 1 / n_components.

In [1], this is called alpha.

topic_word_priorfloat, default=None

Prior of topic word distribution beta. If the value is None, defaults to 1 / n_components. In [1], this is called eta.

output adalah :

- proporsi topik pada dokumen

- distribusi kata pada topik

```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df_tf = pd.read_excel("/content/drive/MyDrive/ppw/tugas/data-tugas/df_tf.xlsx")
df_tf
```


Diketahui dan diperintahkan bahwa harus,

**k=3 alpHa=0.1 beta=0.2**

Kita implementasikan di library LDA dengan,

n_components=3,  topic_word_prior=0.1, doc_topic_prior=0.2

```python
lda = LatentDirichletAllocation(n_components=3, doc_topic_prior=0.2, topic_word_prior=0.1,random_state=42,max_iter=1)
lda_top=lda.fit_transform(df_tf)
```

```python
print(lda_top.shape)  # (no_of_doc,no_of_topics)
print(lda_top)
```

```python
#bobot setiap topik terhadap dokumen
bobot = pd.DataFrame(lda_top, columns=['Topik 1','Topik 2','Topik 3'])
bobot
```

```python
print(lda.components_)
print(lda.components_.shape)  # (no_of_topics*no_of_words)
```

```python
#bobot setiap kata terhadap topik
label=[]
for i in range (1,(lda.components_.shape[1]+1)):
  masukan = df_tf.columns[i-1]
  label.append(masukan)
VT_tabel = pd.DataFrame(lda.components_,columns=label)
VT_tabel.rename(index={0:"Topik 1",1:"Topik 2",2:"Topik 3"}).transpose()
```
